<h1>6. Adding HDI Category Column:</h1>

```python

# Create HDI Category column
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"

hdi_2022_df["HDI Category"] = hdi_2022_df["hdi"].apply(hdi_category)
hdi_2022_df
```

<h2>3. Outlier Detection:</h2>

<h3> Detect outliers in HDI and GNI per Capita using the 1.5 Ã—IQR rule.</h3>

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

south_asia_df["HDI_Outlier"] = detect_outliers(south_asia_df["hdi"])
south_asia_df["GNI_Outlier"] = detect_outliers(south_asia_df["gross_inc_percap"])
south_asia_df["Outlier"] = south_asia_df["HDI_Outlier"] | south_asia_df["GNI_Outlier"]
```

 Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=south_asia_df,
    x="gross_inc_percap",
    y="hdi",
    hue="Outlier",
    palette={True: "red", False: "blue"}
)
plt.title("HDI vs GNI per Capita (South Asia)")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend(title="Outlier")
plt.show()
```

verifiying plot data

```python
# Count outliers
south_asia_df["HDI_Outlier"].value_counts(), south_asia_df["GNI_Outlier"].value_counts()
```

No countries in South Asia were identified as outliers in HDI or GNI per Capita when applying the 1.5Ã—IQR rule. This indicates that the development indicators across the region are relatively consistent, with no country exhibiting extremely high or low values compared to the rest. Even countries such as Maldives and Afghanistan, which represent the highest and lowest values respectively, still fall within the expected statistical range defined by the interquartile boundaries.

The absence of outliers suggests that differences in development levels among South Asian countries are gradual rather than extreme. This may be due to shared regional economic and social characteristics, as well as similar development challenges. As a result, while variations in HDI and GNI per Capita exist, none are sufficiently extreme to be classified as statistical outliers under the IQR-based method.

```python
# Function to detect outliers using IQR
def detect_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return (series < lower) | (series > upper)

# Detect outliers in HDI and GNI per capita
south_asia_df["Outlier"] = detect_outliers(south_asia_df["hdi"]) | detect_outliers(
    south_asia_df["gross_inc_percap"]
)

middle_east_df["Outlier"] = detect_outliers(middle_east_df["hdi"]) | detect_outliers(
    middle_east_df["gross_inc_percap"]
)

# Scatter plots highlighting outliers
plt.figure(figsize=(6,5))
sns.scatterplot(
    data=south_asia_df,
    x="gross_inc_percap",
    y="hdi",
    hue="Outlier"
)
plt.title("South Asia: HDI vs GNI per Capita (Outliers)")
plt.show()

plt.figure(figsize=(6,5))
sns.scatterplot(
    data=middle_east_df,
    x="gross_inc_percap",
    y="hdi",
    hue="Outlier"
)
plt.title("Middle East: HDI vs GNI per Capita (Outliers)")
plt.show()
```
